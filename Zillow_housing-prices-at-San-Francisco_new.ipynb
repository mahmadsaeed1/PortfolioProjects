```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read ZORI
df = pd.read_csv("Metro_zori_sm_month.csv")
df = df[df['RegionID'] == 395057]
df = df.melt(id_vars = ['RegionID','SizeRank','RegionName','RegionType','StateName'])
df.columns = ['RegionID','SizeRank','RegionName','RegionType','StateName','Date','ZORI']
```

```python
new_listings = pd.read_csv("Metro_new_listings_uc_sfrcondo_sm_month.csv")
new_listings = new_listings[new_listings['RegionID'] == 395057]
new_listings = new_listings.melt(id_vars = ['RegionID','SizeRank','RegionName','RegionType','StateName'])
new_listings.columns = ['RegionID','SizeRank','RegionName','RegionType','StateName','Date','New_Listings']
```

```python
new_pending_listings = pd.read_csv("Metro_new_pending_uc_sfrcondo_month.csv")
new_pending_listings = new_pending_listings[new_pending_listings['RegionID'] == 395057]
new_pending_listings = new_pending_listings.melt(id_vars = ['RegionID','SizeRank','RegionName','RegionType','StateName'])
new_pending_listings.columns = ['RegionID','SizeRank','RegionName','RegionType','StateName','Date','New_Pending_Listings']
```
